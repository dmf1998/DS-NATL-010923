### 3b: What are some of the things we could do to deal with null values?
 

Answer:

- We have the options of dropping or imputing the data.
- Because this dataset is not too large, it may not be a good idea to drop the data, but we need to treat each column separately here
- Paid consists of boolean values, but we are only missing one row. Dropping the row sacrifices a lot of information, so it may be best to replace the value with the mode. Later on a correlation between paid and some other category could be used to impute this data point
- like could be imputed with the mean because 1/500 is not going to have a drastic impact on the distribution of the curve
- share has 4 missing values, making this one more difficult to deal with. Because this is a continuous numeric column, a mean or median could be suitable replacements for the values, but if we were to do correlation analysis later on, this might affect the correlations. Dropping these rows might be best, but it is certainly not worth removing the whole column

## Task 12: Navigate the dictionary

## Task 13: Loop to List

Create a list of **track names** of all twenty songs using a for loop or list comprehension.

## Task 14: Create a new dictionary

Create a dictionary called `song_dictionary` which consists of each track name `string` as a key and a `tuple` of artists associated with each track as a value.

```python
# Your code here:
song_dictionary = {song['name']:tuple([artist['name'] for artist in song['artists']]) for song in first_twenty_songs}
song_dictionary
```

## Task 15: Write a function

Create a function with takes an **artist name** and the **song_dictionary** as arguments, and returns a `list` of songs written by that artist. 

```python
# Easier to do things outside of a function first
# Let's try for The Weeknd
for key, val in song_dictionary.items():
    if 'The Weeknd' in val:
        print(key)
```

```python
# Your code here

def find_song_by_artist(artist_name, song_dict):
    
    '''
    Parameters:
    arist_name: a string of an artist's name to be used to search the dictionary
    song_dict:  a dictionary of top_twenty songs with song name as keys and a list of 
    artist names as values
    
    Returns:
    A list of songs which the given artist appeared on
    '''
    
    artist_song_list = []
    for key, val in song_dict.items():
        if artist_name in val:
            artist_song_list.append(key)
    return artist_song_list
    
```
